```python

import cv2
import matplotlib.pyplot as plt
import numpy as np
from collections import namedtuple
import easyocr
from transformers import TrOCRProcessor, VisionEncoderDecoderModel
from PIL import Image

# Define a namedtuple to store text detection results
TextDetectionResult = namedtuple('TextDetectionResult', ['startX', 'startY', 'endX', 'endY'])

def detect_text(image):
    """
    Detects text regions in the image using EasyOCR.

    Args:
        image: The input image (BGR format).

    Returns:
        A list of TextDetectionResult objects containing bounding boxes for detected text regions.
    """
    reader = easyocr.Reader(['en'], gpu=False)  # Load EasyOCR reader for English detection
    boxes = []

    # Perform text detection using EasyOCR and extract bounding boxes
    for detection in reader.readtext(image):
        try:
            top_left, bottom_right = tuple(detection[0][0]), tuple(detection[0][2])
            boxes.append(TextDetectionResult(top_left[0], top_left[1], bottom_right[0], bottom_right[1]))
        except Exception as e:
            print(f"Error parsing EasyOCR detection output: {e}")

    return boxes

def recognize_text(image, boxes, processor, model):
    """
    Performs text recognition on an image region using TrOCR.

    Args:
        image: The input image (BGR format).
        boxes: A list of TextDetectionResult objects containing bounding box coordinates.
        processor: The TrOCR processor.
        model: The TrOCR model.

    Returns:
        A list of recognized text strings.
    """
    recognized_texts = []
    kernel = np.array([[0, -1, 0],
                       [-1, 5,-1],
                       [0, -1, 0]])
    for box in boxes:
        try:
            startX, startY, endX, endY = box
            startX = int(startX)
            startY = int(startY)
            endX = int(endX)
            endY = int(endY)
            print(f"startX: {startX}, startY: {startY}, endX: {endX}, endY: {endY}")
            text_region = image[startY:endY, startX:endX]
            text_region = cv2.cvtColor(text_region, cv2.COLOR_BGR2RGB)
            text_region = cv2.filter2D(image, -1, kernel)
            # text_region = Image.fromarray(text_region)

            pixel_values = processor(images=text_region, return_tensors="pt").pixel_values.to("cuda")
            generated_ids = model.generate(pixel_values)
            recognized_text = processor.batch_decode(generated_ids, skip_special_tokens=True)
            recognized_texts.append(recognized_text)
        except Exception as e:
            print(f"Error processing text region: {e}")
            recognized_texts.append('')

    return recognized_texts

def draw_boxes_and_text(image, boxes, recognized_texts):
    """
    Draws bounding boxes around detected text and overlays recognized text on the image.

    Args:
        image: The input image.
        boxes: A list of TextDetectionResult objects containing bounding boxes.
        recognized_texts: A list of recognized text strings.

    Returns:
        The image with bounding boxes and text overlay.
    """
    font = cv2.FONT_HERSHEY_SIMPLEX
    text_color = (255, 0, 0)  # Red for text overlay
    fimage = np.array(image.copy())
    for i, box in enumerate(boxes):

        if i < len(recognized_texts) and recognized_texts[i] is not None:
            (startX, startY, endX, endY) = box
            text = str(recognized_texts[i])
            print(text)
            cv2.rectangle(fimage, (int(startX), int(startY)), (int(endX), int(endY)), (0, 255, 0), 1)
            cv2.putText(fimage, text, (int(startX), int(startY) + 5), font, 0.65, text_color, 2, cv2.LINE_AA)

    return fimage

def main():

    # Load the image
    image_paths = [
        '/content/Screenshot 2024-06-04 103804.png',
        '/content/Screenshot 2024-06-04 103814.png',
        '/content/Screenshot 2024-06-04 103846.png',
        '/content/Screenshot 2024-06-04 103940.png',
        '/content/Screenshot 2024-06-04 103948.png',
        '/content/Screenshot 2024-06-04 103958.png',
        '/content/Screenshot 2024-06-04 104008.png',
        '/content/Screenshot 2024-06-04 104020.png',
        '/content/Screenshot 2024-06-04 104029.png',
        '/content/Screenshot 2024-06-04 104045.png'
    ]
    for image_path in image_paths:
        image = cv2.imread(image_path)
        processor = TrOCRProcessor.from_pretrained('microsoft/trocr-large-handwritten')
        # model = VisionEncoderDecoderModel.from_pretrained('/content/fine-tuned-trocr')
        model = VisionEncoderDecoderModel.from_pretrained('microsoft/trocr-large-handwritten')
        model.to("cuda")

        # Detect text regions
        boxes = detect_text(image.copy())

        # Recognize text within regions
        recognized_texts = recognize_text(image.copy(), boxes, processor, model)

        # Draw bounding boxes and recognized text on the image
        result_image = draw_boxes_and_text(image.copy(), boxes, recognized_texts)

        # Display results
        plt.figure(figsize=(10 ,20))
        plt.imshow(result_image)
        plt.show()

if __name__ == "__main__":
    main()
```
